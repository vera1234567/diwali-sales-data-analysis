```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('Diwali Sales Data.csv',encoding=('ISO-8859-1'))
df.dropna(inplace=True)
df['Amount']=df['Amount'].astype('int')
sales=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_df=sns.barplot(x='Gender',y='Amount',data=sales)
for bar in sales_df.containers:
    sales_df.bar_label(bar)
    
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('Diwali Sales Data.csv',encoding=('ISO-8859-1'))
df.dropna(inplace=True)
df['Amount']=df['Amount'].astype('int')
state_sales=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
stateplot=sns.barplot(x='State',y='Orders',data=state_sales)
sns.set(rc={'figure.figsize':(19,6)})
for bar in stateplot.containers:
    stateplot.bar_label(bar)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('Diwali Sales Data.csv',encoding=('ISO-8859-1'))
df.dropna(inplace=True)
df['Amount']=df['Amount'].astype('int')
sales_occ=df.groupby(['Occupation','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sales_occ,x='Occupation',y='Amount',hue='Gender')
```

here also maximum shopping is done by females only in those particular sector, which also shows that males are less likely to
do purchases, only purchasing in IT and healthcare sector

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('Diwali Sales Data.csv',encoding=('ISO-8859-1'))
df.dropna(inplace=True)
df['Amount']=df['Amount'].astype('int')
prod_amount=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(34,12)})
sns.barplot(data=prod_amount,x='Product_Category',y='Amount')
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('Diwali Sales Data.csv',encoding=('ISO-8859-1'))
df.dropna(inplace=True)
df['Amount']=df['Amount'].astype('int')
prodid=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_ID',y='Orders',data=prodid)
```

hence married women of age 25-36 of maharashtra, U.P and karnataka working in IT, Healthcare and Aviation have done more shopping, precisely buying more of Food, electronics and Clothing category
